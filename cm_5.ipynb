# Domestic Tourist Arrivals in Uttarakhand (2016-2020)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("data.csv")
df
```

```python
# Line chart for Domestic Tourists
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Domestic Tourists'], marker='o', linestyle='-', color='blue', label='Domestic Tourists')
plt.xlabel('Year')
plt.ylabel('Number of Tourists (Millions)')
plt.title('Domestic Tourist Arrivals in Uttarakhand (2016-2020)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Line chart for Foreign Tourists
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Foreign Tourists'], marker='o', linestyle='-', color='orange', label='Foreign Tourists')
plt.xlabel('Year')
plt.ylabel('Number of Tourists (Millions)')
plt.title('Foreign Tourist Arrivals in Uttarakhand (2016-2020)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Bar chart for Total Tourist Arrivals
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Domestic Tourists'] + df['Foreign Tourists'], color=['blue', 'orange'])
plt.xlabel('Year')
plt.ylabel('Number of Tourists (Millions)')
plt.title('Total Tourist Arrivals in Uttarakhand (2016-2020)')
plt.legend()
plt.show()
```

```python
sns.scatterplot(data=df,x ="Year", y="Domestic Tourists",hue="Year")
```

```python
plt.plot( 'Year', 'Domestic Tourists', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'Year', 'Foreign Tourists', data=df, marker='*', color='olive', linewidth=2)
plt.legend()
plt.show()
```
